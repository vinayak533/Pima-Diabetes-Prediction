# üß† Pima Diabetes Prediction

This project uses a dataset of medical diagnostic measurements to predict whether a patient is likely to have diabetes. The data was originally collected from women of Pima Indian heritage aged 21 and older. We use Python and scikit-learn to explore, visualize, and build a machine learning model for prediction.

## üîç Problem Statement
To predict the presence of diabetes based on features like:
- Number of pregnancies
- Glucose level
- Blood pressure
- Skin thickness
- Insulin level
- BMI
- Diabetes pedigree function
- Age

## ‚úÖ Goals
- Clean and preprocess the dataset.
- Perform exploratory data analysis (EDA).
- Build and evaluate a machine learning classification model.
- Provide insights from model performance.

```python
import numpy as np
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_data = pd.read_csv('diabetes.csv')
diabetes_data 
```

```python
x = diabetes_data.drop(columns = 'Outcome' , axis = 1)
y = diabetes_data['Outcome']
```

```python
print(x)
```

```python
print(y)
```

```python
scaler = StandardScaler()
scaler.fit(x)
standardized_data = scaler.transform(x)
print(standardized_data)
```

```python
x = standardized_data
y = diabetes_data['Outcome']
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,stratify = y,random_state = 2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
classifier = svm.SVC(kernel = 'linear')
classifier.fit(x_train,y_train)
```

```python
x_train_prediction = classifier.predict(x_train)
x_train_accuracy = accuracy_score(x_train_prediction , y_train)
print('Training Accuracy :' , x_train_accuracy)
```

```python
x_test_prediction = classifier.predict(x_test)
x_test_accuracy = accuracy_score(x_test_prediction , y_test)
print('Testing Accuracy :' , x_test_accuracy)
```

```python
input_data = ( 8, 183, 64, 0, 0, 23.3, 0.672, 32)
input_data_np = np.asarray(input_data)
input_data_reshaped  = input_data_np.reshape(1,-1)
```
